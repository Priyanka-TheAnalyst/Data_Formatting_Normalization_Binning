```python
# NORMALIZATION 

# SIMPLE FEATURE SCALING (METHOD 1) 

for i in x:
    df1[i]=df1[i]/df1[i].max()
df1
```

```python
df2=df.copy()
```

```python
# MIN - MAX 
for i in x:
    df2[i]=(df2[i]-df2[i].min())/(df2[i].max()-df2[i].min())
df2
```

```python
df3=df.copy()
```

```python
# Z - SCORE 
for i in x:
    df3[i]=(df3[i]-df3[i].mean())/df[i].std()
df3
```

```python
# BINNING 
bins=np.linspace(min(df["avg"]),max(df["avg"]),4)
bins
```

```python
group_name=["low","medium","high"]
```

```python
df["avg-binned"]=pd.cut(df["avg"],bins,labels=group_name,include_lowest=True)
```
